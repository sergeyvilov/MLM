```python
human_fasta = data_dir + 'fasta/240_mammals/species/Homo_sapiens.fa'
```

```python
human_utr = defaultdict(str)

with open(human_fasta, 'r') as f:
    for line in f:
        if line.startswith('>'):
            fasta_id = line[1:].split(':')[0]
        else:
            human_utr[fasta_id] += line.rstrip()
```
